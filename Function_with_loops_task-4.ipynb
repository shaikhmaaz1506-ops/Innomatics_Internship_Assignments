```python
'''
Problem 1: Mobile Recharge Validation System
 Problem Statement
Create a function to validate a mobile recharge request.
'''
```

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    
    while True:
        try:
            amount = int(input("Enter recharge amount: ₹"))
            
            # Rule 1: Minimum amount check
            if amount < 50:
                print("Recharge amount must be at least ₹50.")
                continue
            
            # Rule 2: Valid plan check
            if amount not in valid_plans:
                print("Invalid plan selected.")
                print("Available plans:", valid_plans)
                continue
            
            # If all conditions are satisfied
            print("Recharge successful for ₹", amount)
            break
        
        except ValueError:
            print("Please enter a valid numeric amount.")

# Function call
validate_recharge()
```

```python
'''
Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:
-Print Reorder Alert if stock < 15
-Print Stock OK otherwise
'''
```

```python
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product} → Reorder Alert (Stock: {quantity})")
        else:
            print(f"{product} → Stock OK (Stock: {quantity})")


# Example dictionary
inventory = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 25,
    "Monitor": 8,
    "Printer": 15
}

# Function call
check_inventory(inventory)
```

```python
'''
Problem 3: Student Result Processing System
Problem Statement
Create a function that:
-Accepts a list of student marks
-Calculates average marks
-Prints Pass if average ≥ 50 else Fail
'''
```

```python
def process_result(marks):
    if len(marks) == 0:
        print("No marks provided.")
        return
    
    average = sum(marks) / len(marks)
    print("Average Marks:", average)
    
    if average >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")


# Example list of marks
student_marks = [60, 45, 70, 55, 40]

# Function call
process_result(student_marks)
```

```python
'''
Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
-Base fare = ₹50
-₹12 per km
-Peak hour → 25% extra
-Allow retry using while loop
'''
```

```python
def cab_fare_calculator():
    BASE_FARE = 50
    PER_KM_RATE = 12
    
    while True:
        try:
            distance = float(input("Enter distance in km: "))
            
            if distance <= 0:
                print("Distance must be greater than 0.")
                continue
            
            peak = input("Is it peak hour? (yes/no): ").lower()
            
            fare = BASE_FARE + (distance * PER_KM_RATE)
            
            if peak == "yes":
                fare += fare * 0.25   # 25% extra
            
            print(f" Total Fare: ₹{fare}")
            
            retry = input("Do you want to calculate again? (yes/no): ").lower()
            if retry != "yes":
                print("Thank you for using Cab Fare Estimator!")
                break
                
        except ValueError:
            print("Please enter valid numeric distance.")

# Function call
cab_fare_calculator()
```

```python
'''
Problem 5: Employee Attendance Eligibility Checker
Problem Statement
Create a function that:
-Accepts attendance list ("P", "A")
-Calculates attendance percentage
-Returns Eligible if ≥ 75% else Not Eligible
'''
```

```python
def check_attendance(attendance_list):
    if len(attendance_list) == 0:
        print("No attendance data available.")
        return
    
    total_days = len(attendance_list)
    present_days = attendance_list.count("P")
    
    percentage = (present_days / total_days) * 100
    
    print(f"Attendance Percentage: {percentage:.2f}%")
    
    if percentage >= 75:
        print("Status: Eligible")
    else:
        print("Status: Not Eligible")


# Example attendance list
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Function call
check_attendance(attendance)
```

```python
'''
Problem 6: Password Strength Checker
Problem Statement
Create a function to check password strength.
Rules:
-Minimum length = 8
-Must contain at least one digit
-Must contain at least one special character (@#$)
'''
```

```python
def check_password_strength(password):
    special_characters = "@#$"
    
    # Rule 1: Minimum length
    if len(password) < 8:
        print("Weak Password: Minimum length should be 8 characters.")
        return
    
    # Rule 2: At least one digit
    has_digit = any(char.isdigit() for char in password)
    
    # Rule 3: At least one special character
    has_special = any(char in special_characters for char in password)
    
    if not has_digit:
        print("Weak Password: Must contain at least one digit.")
    elif not has_special:
        print("Weak Password: Must contain at least one special character (@#$).")
    else:
        print("Strong Password")
        

# Example usage
pwd = input("Enter your password: ")
check_password_strength(pwd)
```
