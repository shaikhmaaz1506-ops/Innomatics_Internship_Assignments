```python
'''
Problem 1: Unique Words in a Sentence
Problem Statement
Given a sentence, identify all unique words and display:
-The count of unique words
-The unique words themselves
'''
```

```python
# input sentence
sentence = "python is easy and python is powerful"

# split words from sentence
words = sentence.split()

# split() takes only uniques words
unique_words = set(words)

print("unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
'''
Problem 2: Highest Salary from Employee Data
Problem Statement
Store employee names and their salaries, and determine which employee has the highest salary.
'''
```

```python
employee = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employee, key = employee.get)
highest_salary = employee[highest_employee]

print("Highest Salary:",highest_employee,"-",highest_salary)
```

```python
'''
Problem 3: Find Maximum and Minimum Values
Problem Statement
Find the maximum and minimum values from a list of numbers.
'''
```

```python
# Given list
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Print results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
'''
Problem 4: Count Products Above a Price Threshold
Problem Statement
Count how many products have a price greater than 1000.
'''
```

```python
# Given list
prices = [450, 1200, 899, 1500, 300]

# Initialize counter
count = 0

# Loop through prices
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

```python
'''
Problem 5: Calculate Attendance Percentage
Problem Statement
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).
'''
```

```python
# Given list
attendance = ["P", "P", "A", "P", "P"]

# Count total days
total_days = len(attendance)

# Count number of Present days
present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

# Calculate percentage
attendance_percentage = (present_count / total_days) * 100

# Print result
print("Attendance Percentage:", attendance_percentage)
```

```python
'''
Problem 7: Count Character Frequency
Problem Statement
Count how many times each character appears in a given string.
'''
```

```python
# Given string
text = "pythonp"

# Create empty dictionary
char_count = {}

# Loop through each character
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

# Print result
print(char_count)
```

```python
'''
Problem 10: Calculate Average Marks
Problem Statement
Calculate the average marks of a student from a list of marks.
'''
```

```python
# Given list of marks
marks = [70, 75, 80, 65]

# Calculate total marks
total = sum(marks)

# Calculate number of subjects
count = len(marks)

# Calculate average
average = total / count

# Print result
print("Average Marks:", average)
```
