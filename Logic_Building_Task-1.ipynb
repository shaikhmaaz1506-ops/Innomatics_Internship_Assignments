```python
'''
1) User Login Check
Problem Statement:
Given a username and password, check whether login is successful.
'''
```

```python
# correct username and password
username = "admin"
password = "1234"

# input from user
input_username = input("Enter a username")
input_password = input("Enter password")

if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
'''
2) Pass / Fail Analyzer
Problem Statement:
Analyze student results.
'''
```

```python
marks = [45, 78, 90, 33, 60]

# variable for storing count
pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
'''
3) Simple Data Cleaner
Problem Statement:
Clean and standardize user names.
'''
```

```python
names = [" Alice ", "bob", " CHARLIE "]

# Empty list 
cleaned_list = []

for name in names:
    cleaned_names = name.strip().lower()  # removing space and converting to lower case
    cleaned_list.append(cleaned_names)    # Add cleaned names to empty list

print("Cleaned List =", cleaned_list)
```

```python
'''
4) Message Length Analyzer
Problem Statement:
Analyze message sizes.
'''
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)  # len() returns length of a string
    print("message:", msg)
    print("Length:", length)

    if length > 10:
        print("This message is longer than 10 characters")    
```

```python
'''
5) Error Message Detector
Problem Statement:
Detect error messages from system logs.
'''
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# variable for storing values of error
error_count = 0

for i in logs:
    if i == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
